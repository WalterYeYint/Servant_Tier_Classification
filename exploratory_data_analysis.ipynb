## **QUESTIONS**

### **Conditional Selection**

In performing exploratory data analysis, it is important to be able to select subsets of data to perform analysis or comparisons.

**Which servant has the highest min HP?**
Here, we will return the entire row.

```python
df[df.min_HP == df.min_HP.max()]
```

We will return specific column values.

Further question, what is the servant's individual info? 

```python
HPServant = df[df.min_HP == df.min_HP.max()]
HPServant.name
```

```python
HPServant.star_absorption
```

```python
HPServant.NP_charge_attack
```

**Which servant has max_attack more than 13000?**

```python
df[df.max_attack > 13000]
```

### **GroupBy() function**

**Which servant in the Archer class has the highest max_attack?**

```python
Archers = df.groupby('class').get_group('Archer')
```

```python
Archers[Archers.max_attack == Archers.max_attack.max()]
```

**Of the 5 positions, which position scores the most points?**

We first group servants by their class

```python
pd.set_option('display.max_rows', None)
df.groupby('class').max_attack.describe()
```

### **Histograms**

We'll also try to answer this question by showing some histogram plots. So, to make it a bit easier, let's create a subset dataframe.

```python
MAX_ATTACK = df[['class','max_attack']]

classes = ['Saber','Archer','Lancer','Rider','Caster']
MAX_ATTACK = MAX_ATTACK[ MAX_ATTACK['class'].isin(classes)  ]

MAX_ATTACK
```

#### **pandas built-in visualization**

```python
MAX_ATTACK['max_attack'].hist(by=MAX_ATTACK['class'])
```

```python
MAX_ATTACK['max_attack'].hist(by=MAX_ATTACK['class'], layout=(1,5))
```

```python
MAX_ATTACK['max_attack'].hist(by=MAX_ATTACK['class'], layout=(1,5), figsize=(16,2))
```

#### **Seaborn data visualization**

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(MAX_ATTACK, col="class")
g.map(plt.hist, "max_attack");
```

### **Heat map**

#### Compute the correlation matrix

```python
corr = df.corr()
corr
```

#### Make the heat map

```python
sns.heatmap(corr)
```

#### Adjust figure size of heat map

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(7,5))
sns.heatmap(corr, square=True)
```

#### Mask diagonal half of heat map (Diagonal correlation matrix)

```python
# https://seaborn.pydata.org/generated/seaborn.heatmap.html

import numpy as np
import seaborn as sns

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr, mask=mask, vmax=1, square=True)
```
